## Split The Train DataSet into Training and Testing DataSet

```python
# import the train_test_split function from the sklearn library
from sklearn.model_selection import train_test_split
```

```python
# Apply the function and split the data 
x_train,x_test,y_train,y_test = train_test_split(Train.iloc[:,:10],
                                                 Train['class'],test_size = 0.25)
```

```python
# for 1 hot encoding
y_train = pd.get_dummies(y_train,columns=['class'])
y_test = pd.get_dummies(y_test,columns=['class'])
```

## Create The Neural Networking Model and Fit it on The DataSet  

```python
# import keras and other layers
 
import keras 
from keras.layers import Dense, Dropout
from keras.models import Sequential
```

```python
# model architecture

model=Sequential()

model.add(Dense(3600,activation='relu',input_shape=(10,)))
model.add(Dense(2700,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(2500,activation='relu'))
model.add(Dense(2000,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(2700,activation='relu'))
model.add(Dense(1800,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(900,activation='relu'))
model.add(Dense(800,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(600,activation='relu'))
model.add(Dense(500,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(400,activation='relu'))
model.add(Dense(300,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(250,activation='relu'))
model.add(Dense(12,activation='softmax'))

# model compilation

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])

# fitting the model

model.fit(x_train,y_train,epochs=100)
```

## Check The Models Accuracy

```python
# Predict on x_test dataset 
prediction = model.predict(x_test)
prediction.shape
```

```python
# get the predicted class value
predicted_class = np.argmax(prediction, axis = -1)
actual_class = np.argmax(np.array(y_test), axis = -1)
```

```python
predicted_class
```

```python
actual_class
```

```python
# get the accuracy of the model
per = len(list(x_test))
total = len(list(x_test))
for i in range(per):
    if predicted_class[i] != actual_class[i]:
        per -= 1
print(round(per/total,2))
```

## Now Predict The Class of Test DataSet Using The Model

```python
# Now Predict the Test Datset "class" field
result = model.predict(Test.iloc[:,:10])
result.shape
```

```python
# get the class value
result = np.argmax(result, axis = -1)
```

```python
# Index 
classes = range(12)

# Lables Name
lables = ['A','B','C','D','E','F','G','H','I','W','X','Y']

# Zip the index and there corresponding lable name to make a dictionary
class_lable = dict(zip(classes, lables))
```

```python
# now replace the index value with it's class value
final_result = []
for n in range(len(result)):
    final_result.append(class_lable[result[n]])
```

```python
# Save the prediction as result.csv file in the local directory
final_result = pd.DataFrame(final_result)
final_result.to_csv('result.csv')
```
